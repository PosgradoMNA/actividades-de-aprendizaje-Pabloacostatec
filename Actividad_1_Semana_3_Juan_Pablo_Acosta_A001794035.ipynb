```python
#Define DB link
input = "https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv"
```

```python
#Import pandas an use pandas to import db
import pandas as pd
import numpy as np

InitialDF = pd.read_csv(input)
InitialDF
```

```python
#Verify that there are no lost values, this outputs a boolean for each column true in case they got any null value

InitialDF.isnull().any()
```

```python
#Count how many null values are in each column

InitialDF.isna().sum().sum()
```

```python
#Drop the rows that overpass the treshold of null values, if you use the "axis" variable, it indicates it will apply the treshold to the column
ndf = InitialDF.copy()
ndf.dropna(thresh=3,inplace = True)#,axis = 1
ndf
```

```python
#Get the indexes of the rows that got nulls for all columns

HeadersList = list(InitialDF.columns) #Gets a directory of all headers
HeadersList

AllRowsArray = []

for i in range(25): #This loop appends the rows that have a null value into the 'AllRowsArray
  MissingValuesIndex = np.where(InitialDF[HeadersList[i]].isnull())[0]
  AllRowsArray= np.append(AllRowsArray,MissingValuesIndex)
 # print(MissingValuesIndex)

from collections import Counter #This library lets me use AllRowsKeys

IndexesNumber = Counter(AllRowsArray).keys() # equals to list(set(words))
CountOfNulls= Counter(AllRowsArray).values() # counts the elements' frequency

RowsNullCounter = np.array([[IndexesNumber],[CountOfNulls]]) #Creates the final matrix that counts the index that has nulls and the amount of columns that has nulls

RowsNullCounter


#print(AllRowsArray)
#Get the in descending order all the rows that got null values


#b = InitialDF.iloc[2,5]
#b  
```
